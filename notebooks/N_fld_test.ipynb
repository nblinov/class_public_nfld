```python
params = {
    'output': 'mPk,mTk,vTk',#'mPk',
    'A_s': 2.3e-9,
    'n_s': 0.9624, 
    '100*theta_s':1.042143,
    'omega_b': 0.02226,
    'omega_cdm': 0.1186,
    # other options and settings
    'compute damping scale':'yes',#, # needed to output the time of damping scale crossing
    'gauge':'synchronous'
    #'gauge':'newtonian'
}

def get_z_eq(bg):
    #print bg
    zs = bg['z']
    rho_m = bg['(.)rho_cdm']+bg['(.)rho_b']
    rho_r = bg['(.)rho_ur']+bg['(.)rho_g'] + 1e-100
    if '(.)rho_ur_fld' in bg:
        rho_m = rho_m + bg['(.)rho_ur_fld']
    if '(.)rho_dr' in bg:
        rho_r = rho_r + bg['(.)rho_dr']
    """
    plt.loglog(1/(1.+bg['z']),rho_m)
    plt.loglog(1/(1.+bg['z']),rho_r)
    plt.xlim([1e-6,1e-3])
    plt.ylim([1,1e12])
    plt.show()
    """
    rho_ratio = rho_m/rho_r
    root_func = interpolate.interp1d(np.log10(zs),np.log10(rho_ratio),kind='linear')
    zeq = np.power(10.,optimize.brentq(root_func,np.log10(100.),np.log10(1e5)))
    #print zeq, root_func(np.log10(zeq))
    return zeq

"""
Given a desired redshift of 
equality, return the new DM abundance needed to achieve this
"""
def fix_z_eq_from_cdm(para, z_eq):
    M = Class()
    M.set(para)   
    def get_zeq_from_cdm(omcdm):
        M.set(para)
        M.set({'omega_cdm':omcdm})
        M.compute(['background'])
        z_eq_computed = get_z_eq(M.get_background())#M.z_eq()
        #print M.pars
        M.struct_cleanup()  # clean output
        M.empty() 
        return z_eq_computed - z_eq
    
    #print get_zeq_from_cdm(params['omega_cdm']/1.3), "\t", get_zeq_from_cdm(params['omega_cdm']*1.3)
    return optimize.brentq(get_zeq_from_cdm, params['omega_cdm']/1.3,params['omega_cdm']*1.3,xtol=1e-3)#

"""
Same as above, but hold the 
baryon-to-dm ratio fixed
"""
def fix_z_eq_from_cdm_at_fixed_b_frac(para, z_eq, omegab_over_omegacdm=0.18769):
    M = Class()
    M.set(para)   
    def get_zeq_from_cdm(omcdm):
        M.set(para)
        M.set({'omega_cdm':omcdm})
        M.set({'omega_b': omegab_over_omegacdm*omcdm})
        M.compute(['background'])
        z_eq_computed = get_z_eq(M.get_background())#M.z_eq()
        #print M.pars
        M.struct_cleanup()  # clean output
        M.empty() 
        return z_eq_computed - z_eq
    
    #print get_zeq_from_cdm(params['omega_cdm']/1.3), "\t", get_zeq_from_cdm(params['omega_cdm']*1.3)
    return optimize.brentq(get_zeq_from_cdm, params['omega_cdm']/1.3,params['omega_cdm']*1.3,xtol=1e-3)#

def get_k_mode_model(k, base_param, fix_zeq=True, fix_rd=False):
    # adjust parameters to fix zeq to this value in every model
    zeq_common = 3365.

    # Run Lambda CDM
    para = copy.deepcopy(base_param)
    if fix_zeq:
        omega_cdm = fix_z_eq_from_cdm(para,zeq_common)
        para['omega_cdm'] = omega_cdm  
    
    if fix_rd:
        Yp_new = get_new_Yp_from_params(para)
        print "Adjusted Yp to cancel effect on diffusion = ", Yp_new
        para['YHe'] = Yp_new
    
    para['k_output_values'] = k

    M = Class()
    M.set(para)
    M.compute()    
    
    return M

def plot_k_mode_evolution(k, N_fld):
    para_new = copy.deepcopy(params)
    para_new['N_fld'] = N_fld


    M = get_k_mode_model(k, para_new)

    bg = M.get_background()
    a = 1./(1.+bg['z'])
    rho_rad = bg['(.)rho_ur'] + bg['(.)rho_ur_fld'] + bg['(.)rho_g']
    aeq = 1./(1.+get_z_eq(bg))
    DNeff = bg['(.)rho_ur_fld']/(bg['(.)rho_ur']/3.046)
    print "DNeff at equality = ",DNeff

    all_pert = M.get_perturbations()
    print all_pert['scalar'][0].keys()

    x = all_pert['scalar'][0]['a']
    x = k*all_pert['scalar'][0]['tau [Mpc]']

    plt.plot(x,-all_pert['scalar'][0]['delta_cdm'],label=r'$\delta_{cdm}$')
    plt.plot(x,np.fabs(all_pert['scalar'][0]['delta_ur']),label=r'$\delta_{ur}$')
    plt.plot(x,np.fabs(all_pert['scalar'][0]['delta_ur_fld']),label=r'$\delta_{fld}$')
    plt.plot(x,np.fabs(all_pert['scalar'][0]['delta_g']),label=r'$\delta_\gamma$')
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel(r'$k\tau$')
    plt.ylabel(r'$\delta_i$')
    plt.show()

    plt.plot(x,-all_pert['scalar'][0]['theta_cdm'],label=r'$\theta_{cdm}$')
    plt.plot(x,np.fabs(all_pert['scalar'][0]['theta_ur']),label=r'$\theta_{ur}$')
    plt.plot(x,np.fabs(all_pert['scalar'][0]['theta_ur_fld']),label=r'$\theta_{fld}$')
    plt.plot(x,np.fabs(all_pert['scalar'][0]['theta_g']),label=r'$\theta_\gamma$')
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel(r'$k\tau$')
    plt.ylabel(r'$\theta_i$')
    plt.show()
```

```python
params = {
    'output': 'tCl lCl',
    'l_max_scalars': 2500,
    'lensing': 'yes',
    'A_s': 2.3e-9,
    'n_s': 0.9624, 
    '100*theta_s':1.0411,
    'omega_b': 0.02226,
    'omega_cdm': 0.1186,
    'compute damping scale':'yes'
}

# compare models with N_ur vs models with Nfld
def make_neff_vs_nfld_plot(Nur, Nfld):
    # Create an instance of the CLASS wrapper
    M_ur = Class()
    M_ur.set(params)
    M_ur.set({'N_ur': Nur})
    M_ur.compute()
    
    M_fld = Class()
    M_fld.set(params)
    M_fld.set({'N_fld': Nfld})
    M_fld.compute()

       
    # Evolution of the Cls
    f, ax = plt.subplots(1,2,figsize=(12.,4.))
    plt.title('Power Spectrum')
    
    ells = M_ur.lensed_cl(2500)['ell']
    ax[0].plot(ells,ells*(ells+1)*M_ur.lensed_cl(2500)['tt']/(2.*np.pi),color='black',lw=1,label=r'$N_{ur}$')
    ax[0].plot(ells,ells*(ells+1)*M_fld.lensed_cl(2500)['tt']/(2.*np.pi),color='green',lw=1,label=r'$N_{fld}$')
    ax[0].plot(ells_SM,ells_SM*(ells_SM+1)*M_SM.lensed_cl(2500)['tt']/(2.*np.pi),':',color='red',label=r'$\Lambda$CDM')
    ax[0].set_xscale('log')
    ax[0].set_xlim([2,2500])
    ax[0].set_ylabel(r'$C_\ell$')
    ax[0].legend(loc='upper left')

    ax[1].plot(ells,(M_ur.lensed_cl(2500)['tt']-M_SM.lensed_cl(2500)['tt'])/(1e-20+M_SM.lensed_cl(2500)['tt']),color='black',lw=1)
    ax[1].plot(ells,(M_fld.lensed_cl(2500)['tt']-M_SM.lensed_cl(2500)['tt'])/(1e-20+M_SM.lensed_cl(2500)['tt']),color='green',lw=1)
    ax[1].set_ylabel(r'$\Delta C_\ell/C_\ell^{\Lambda\mathrm{CDM}}$')
    ax[1].set_xlim([2,2500])
    ax[1].set_xscale('log')
    plt.show()
    
# Compare models with both Nur and Nfld against LCDM   
def make_neff_and_nfld_plot(Nur, Nfld):
    # Create an instance of the CLASS wrapper
    M = Class()
    M.set(params)
    M.set({'N_ur': Nur, 'N_fld': Nfld})
    M.compute()
       
    # Evolution of the Cls
    f, ax = plt.subplots(1,2,figsize=(12.,4.))
    plt.title('Power Spectrum')
    
    ells = M.lensed_cl(2500)['ell']
    ax[0].plot(ells,ells*(ells+1)*M.lensed_cl(2500)['tt']/(2.*np.pi),color='black',lw=1,label=r'$N_{ur}+N_{fld}$')
    ax[0].plot(ells_SM,ells_SM*(ells_SM+1)*M_SM.lensed_cl(2500)['tt']/(2.*np.pi),':',color='red',label=r'$\Lambda$CDM')
    ax[0].set_xscale('log')
    ax[0].set_xlim([2,2500])
    ax[0].set_ylabel(r'$C_\ell$')
    ax[0].legend(loc='upper left')

    ax[1].plot(ells,(M.lensed_cl(2500)['tt']-M_SM.lensed_cl(2500)['tt'])/(1e-20+M_SM.lensed_cl(2500)['tt']),color='black',lw=1)
    ax[1].set_ylabel(r'$\Delta C_\ell/C_\ell^{\Lambda\mathrm{CDM}}$')
    ax[1].set_xlim([2,2500])
    ax[1].set_xscale('log')
    plt.show()
    
```
